## Kurtosis

**Kurtosis** is a statistical measure that describes the distribution of data in a dataset.

It provides information about the tails and peaks of the data's distribution. 

In simpler terms, kurtosis tells you about the shape of the probability distribution of a real-valued random variable.

There are different ways to measure kurtosis, but one common method is to compare the distribution to that of a normal distribution, which is known as mesokurtic. 


The kurtosis value can be categorized into three main types:

- Mesokurtic (Kurtosis = 0): The distribution has a similar kurtosis to a normal distribution.
- Leptokurtic (Kurtosis > 0): The distribution has fatter tails and a sharper peak than a normal distribution. This indicates more extreme values.
- Platykurtic (Kurtosis < 0): The distribution has lighter tails and a flatter peak than a normal distribution. This indicates fewer extreme values.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Create a dataset without kurtosis (normal distribution)
data_without_kurtosis = np.random.normal(loc=0, scale=1, size=1000)

# Create a dataset with kurtosis (gamma distribution)
data_with_kurtosis = np.random.gamma(shape=2, scale=1, size=1000)

# Create dataframes
df_without_kurtosis = pd.DataFrame({'Value': data_without_kurtosis, 'Distribution': 'Without Kurtosis'})
df_with_kurtosis = pd.DataFrame({'Value': data_with_kurtosis, 'Distribution': 'With Kurtosis'})

# Combine dataframes
df_combined = pd.concat([df_without_kurtosis, df_with_kurtosis])

# Plot using seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=df_combined, x='Value', hue='Distribution', kde=True)
plt.title('Comparison of Datasets with and without Kurtosis')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot using seaborn - Without Kurtosis
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(data=df_without_kurtosis, x='Value', kde=True, color='blue')
plt.title('Dataset Without Kurtosis')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot using seaborn - With Kurtosis
plt.subplot(1, 2, 2)
sns.histplot(data=df_with_kurtosis, x='Value', kde=True, color='orange')
plt.title('Dataset With Kurtosis')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
